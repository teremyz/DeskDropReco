```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = pd.DataFrame(
    {
        "model": [
            "PopularityM",
            "PopularityM",
            "MatrixFactorization",
            "MatrixFactorization",
            "XGB",
            "XGB",
        ],
        "metric": [
            "PrecisionAt10",
            "RecallAt10",
            "PrecisionAt10",
            "RecallAt10",
            "PrecisionAt10",
            "RecallAt10",
        ],
        "value": [0.0003, 0.001, 0.00228, 0.00998, 0.0008, 0.0084],
    }
)

# Pivot the data for easier plotting
pivot_data = data.pivot(index="metric", columns="model", values="value")

# Plot settings
fig, ax = plt.subplots(figsize=(10, 6))

# Define bar width and positions
bar_width = 0.25
index = range(len(pivot_data))

# Plotting bars for each model
for i, model in enumerate(pivot_data.columns):
    ax.bar(
        [x + i * bar_width for x in index],
        pivot_data[model],
        width=bar_width,
        label=model,
    )

# Adding labels and title
ax.set_xlabel("Metric")
ax.set_ylabel("Metric Value")
ax.set_title("Precision and Recall at 10 for Different Models")
ax.set_xticks([x + bar_width for x in index])
ax.set_xticklabels(pivot_data.index)
ax.legend(title="Model")

plt.show()
```
